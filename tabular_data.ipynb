```python
df[['age', 'potential', 'weight_kgs']].groupby(by=['age'], as_index=False).mean()
```

В результате можем наблюдать следующую закономерность: с увеличением возраста потенциал футболиста падает, а метаболизм ухудшается, что вызывает увеличение веса
