# Ассоциативные правила

```python
import pandas as pd
import numpy as np
from apyori import apriori
```

```python
df = pd.read_csv('datasets/Market_Basket_Optimisation.csv')
df
```

Заполняем пропущенные значения:

```python
df = df.ffill(axis=1)
df
```

Создадим список транзакций, перебирая строки и столбцы и преобразуя значения в string. 

```python
transactions = []
for i in range(0, 7500): 
    transactions.append([str(df.values[i,j]) for j in range(0, 20)])
```

Выполняем ассоциативный анализ правил на транзакциях, установив параметры: 

```python
result = list(apriori(transactions, min_support = 0.003, min_confidence = 0.2, min_lift = 4, min_length = 2))
```

```python
association_results = list(result) #преобразовали вывод алгоритма в список с именем
```

Выводим длину списка:

```python
print(len(result))
```

23 ассоциативных правила на основе анализа данных.

Выводим первый элемент из списка result:

```python
print(result[0])
```

items(предметы): {'light cream', 'chicken'}

Support(поддержка): 0.00453 (это доля транзакций, в которых встречаются 'light cream' и 'chicken')

Ordered_statistics(статистика_порядка): содержит информацию о правиле 'light cream' -> 'chicken'

Confidence(уверенность): 0.2906 (это вероятность, что 'chicken' будет куплено, если куплен 'light cream')

Lift(поддержка): 4.8433 (это показатель, который говорит о том, насколько более вероятно купить 'chicken', если также куплен 'light cream', по сравнению со случаем, когда они покупаются независимо)

Создаем DataFrame с использованием результатов ассоциативного анализа и сохраняем его в переменной nm:

```python
rules=[]
for item in result:

    # первый индекс внутреннего списка
    # Содержит базовый предмет и добавляет предмет
    pair = item[0] 
    items = [x for x in pair]
    v1=str(items[0]) 
    v2= str(items[1])

    #второй индекс внутреннего списка
    v3= str(item[1])

    #третий индекс списка, расположенный на 0-м месте
    #третьего индекса внутреннего списка

    v4= str(item[2][0][2])
    v5=str(item[2][0][3])
    rows=(v1,v2,v3,v4,v5)
    rules.append(rows)
label=['title1','title2','suppot','confidence','lift']  
nm=pd.DataFrame.from_records(rules,columns=label)
nm
```

Сортируем DataFrame nm по столбцу 'lift' в порядке убывания и выводим первые 20 строк. Это позволит увидеть топ-20 ассоциативных правил по показателю поддержки.


```python
nm.sort_values(by=['lift'], ascending=False).head(20)
```

Находим правила, представляющие интерес. Усатановим более жесткие пороги для параметров ассоциативного анализа, такие как confidence(уверенность) и support(поддержка). Увеличим значения min_support и min_confidence.

```python
result = list(apriori(transactions, min_support=0.005,min_confidence=0.3, min_lift=4, min_length=2))
result
```

Из этих правил с товарами 'spaghetti', 'herb & pepper' и 'ground beef' имеет наивысший лифт (4.0038), что может считаться наиболее интересным с точки зрения ассоциации между товарами

Нахождение правила с максимальным лифтом:

```python
max_lift_rule = nm.sort_values(by='lift', ascending=False).iloc[0]
max_lift_rule
```
